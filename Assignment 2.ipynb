```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/SajjadMustafa10/DeepLearning/main/Churn_Modelling.csv")
df.sample(10)
```

```python
df.drop(columns=['RowNumber','CustomerId', 'Surname'],axis='columns',inplace=True)
```

```python
df.dtypes
```

```python
df.EstimatedSalary.values
```

```python
pd.to_numeric(df.EstimatedSalary)
```

```python
def print_unique_col_values(df):
  for column in df:
    if df[column].dtypes=='object':
      print(f'{column}: {df[column].unique()}')
```

```python
print_unique_col_values(df)
```

```python
df['Gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
for col in df:
  print(f'{col}: {df[col].unique()}')
```

```python
df1 = pd.get_dummies(data=df, columns=['Geography'])
df1.columns
```

```python
df1.sample(10)
```

```python
cols_to_scale = ['CreditScore','Balance','EstimatedSalary', 'Age', 'Tenure', 'NumOfProducts']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df1[cols_to_scale] = scaler.fit_transform(df1[cols_to_scale])
```

```python
df1.sample(5)
```

```python
X = df1.drop('Exited',axis='columns')
y = df1['Exited']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
X_train.shape,X_test.shape,X_train[:10]
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(12, input_shape=(12,), activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=150)
```

```python
yp = model.predict(X_test)
yp[:7]
```

```python
y_pred = []
for element in yp:
  if element > 0.5:
    y_pred.append(1)
  else:
    y_pred.append(0)
```

```python
y_pred[:7], y_test[:7]
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test,y_pred))
```
